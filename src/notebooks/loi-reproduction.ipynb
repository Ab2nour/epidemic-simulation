## Histogramme des lois

```python
for name, distribution in distributions.items():
    plt.figure(figsize=(8, 8))
    plt.title(
        f"{name}\nMoyenne = {distribution.mean() : .3f}\nVariance = {distribution.var() : .3f}"
    )
    plot_distribution(distribution)

    filename = name.replace("=", "").replace("/", "")
    plt.savefig(f"data/plots/distribution/histogram/{filename}.png")

    plt.show()
```
